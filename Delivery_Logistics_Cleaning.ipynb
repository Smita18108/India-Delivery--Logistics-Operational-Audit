```python
import pandas as pd
import os

# We use the path from your previous output
path = "/root/.cache/kagglehub/datasets/kundanbedmutha/delivery-logistics-dataset-india-multi-partner/versions/1"
file_path = os.path.join(path, "Delivery_Logistics.csv")

# Load the data
df = pd.read_csv(file_path)

# 1. See the size of the business (How many rows/columns?)
print(f"Dataset Loaded: {df.shape[0]} delivery records found.\n")

# 2. See the column names (The 'Headers' of our business)
print("Business Columns:", df.columns.tolist())

# 3. View the first 5 records
df.head()

```

```python
import pandas as pd
# 1. Convert the object to a proper 'Datetime' object
# Pandas will automatically see these as nanoseconds from 1970-01-01
df['delivery_time_dt'] = pd.to_datetime(df['delivery_time_hours'])
df['expected_time_dt'] = pd.to_datetime(df['expected_time_hours'])
# 2. Extract the 'nanosecond' part.
# In this specific dataset, 1 nanosecond = 1 hour of delivery time.
df['delivery_time_h'] = df['delivery_time_dt'].dt.nanosecond
df['expected_time_h'] = df['expected_time_dt'].dt.nanosecond
# 3. Validation Check
# Let's compare the 'Old' column with our 'New' clean column
print("--- Validation Check ---")
print(df[['delivery_time_hours', 'delivery_time_h']].head())
```

# The Delay Difference Between the delivery_time and the expected_time in Hrs

```python

# 4. Now calculate the actual Business Delay
df['delay_diff'] = df['delivery_time_h'] - df['expected_time_h']
df.head()

```

```python

# See the unique categories and how many deliveries were made for each
print("--- Package Categories Count ---")
print(df['package_type'].value_counts())

print("\n--- Vehicle Fleet Count ---")
print(df['vehicle_type'].value_counts())

```

```python

# Define what we consider "Light Items"
light_items = ['documents', 'pharmacy', 'cosmetics']

# Define "Heavy Vehicles" that consume lots of fuel
heavy_vehicles = ['truck', 'van']

# Identify trips where we used a Heavy Vehicle for a Light Item
money_leaks = df[(df['package_type'].isin(light_items)) & (df['vehicle_type'].isin(heavy_vehicles))]

# Calculate the potential wasted cost
total_wasted_trips = len(money_leaks)
total_wasted_money = money_leaks['delivery_cost'].sum()

print(f"--- Profit Leak Report ---")
print(f"Total 'High-Waste' Trips identified: {total_wasted_trips}")
print(f"Total revenue spent on inefficient fuel trips: ₹{total_wasted_money:,.2f}")

# Breakdown by Region to see where the manager is failing
print("\nWaste Breakdown by Region:")
print(money_leaks['region'].value_counts())

```

```python

# Real 2026 Metrics
diesel_cost_per_km = 5.84   # Diesel Van cost per km
ev_bike_cost_per_km = 1.20  # EV Bike cost per km

# Total Distance of the 2,813 High-Waste trips
# (Assuming average trip distance from your distance_km column)
total_dist = money_leaks['distance_km'].sum()

# 1. What we ACTUALLY spent (Diesel)
actual_diesel_spend = total_dist * diesel_cost_per_km

# 2. What we COULD have spent (EV Bike)
potential_ev_spend = total_dist * ev_bike_cost_per_km

# 3. The Real-World Saving
net_savings = actual_diesel_spend - potential_ev_spend

print(f"--- Real-World Waste Analysis (North Region) ---")
print(f"Total Distance Wasted: {total_dist:,.2f} km")
print(f"Current Actual Spend (Diesel): ₹{actual_diesel_spend:,.2f}")
print(f"Potential Spend (EV Transition): ₹{potential_ev_spend:,.2f}")
print(f"NET PROFIT OPPORTUNITY: ₹{net_savings:,.2f}")

```

```python

# 1. Calculate the 'Real' cost per kilometer for every single trip (25,000 rows)
df['real_cost_per_km'] = df['delivery_cost'] / df['distance_km']

# 2. Create a "Cost Matrix"
# This finds the average cost per km for every combination of Vehicle + Package
cost_matrix = df.groupby(['vehicle_type', 'package_type'])['real_cost_per_km'].mean().unstack()

# 3. View the Matrix
print("--- The Precision Cost Matrix (₹ per KM) ---")
display(cost_matrix)


```

```python

# 1. Identify the 'High-Waste' trips again
light_items = ['documents', 'pharmacy', 'cosmetics']
heavy_vehicles = ['truck', 'van']
money_leaks = df[(df['package_type'].isin(light_items)) & (df['vehicle_type'].isin(heavy_vehicles))].copy()

# 2. Calculate what we ACTUALLY spent on these 2,813 trips
actual_spend = money_leaks['delivery_cost'].sum()

# 3. Calculate what it WOULD have cost if we used an 'EV Bike' for these EXACT trips
# We pull the specific rate for (EV Bike + Package Type) from our cost_matrix
def calculate_optimized_cost(row):
    # Get the specific rate for an EV Bike carrying that specific package type
    optimized_rate = cost_matrix.loc['ev bike', row['package_type']]
    return row['distance_km'] * optimized_rate

money_leaks['optimized_cost'] = money_leaks.apply(calculate_optimized_cost, axis=1)
optimized_total_spend = money_leaks['optimized_cost'].sum()

# 4. Final Business Result
total_savings = actual_spend - optimized_total_spend

print(f"--- 2026 Logistics Optimization Audit ---")
print(f"Total High-Waste Deliveries: {len(money_leaks)}")
print(f"Actual Money Spent (Inefficient): ₹{actual_spend:,.2f}")
print(f"Optimized Money Spent (EV Bike):   ₹{optimized_total_spend:,.2f}")
print(f"NET PROFIT RECOVERED:             ₹{total_savings:,.2f}")


```

```python

# Find the absolute cheapest vehicle for each package type
cheapest_vehicles = cost_matrix.idxmin()
cheapest_rates = cost_matrix.min()

def calculate_best_fit_cost(row):
    # Find the lowest possible rate for this specific package type
    best_rate = cheapest_rates[row['package_type']]
    return row['distance_km'] * best_rate

money_leaks['best_fit_cost'] = money_leaks.apply(calculate_best_fit_cost, axis=1)
best_fit_total = money_leaks['best_fit_cost'].sum()
real_savings = actual_spend - best_fit_total

print(f"--- Final Precision Optimization ---")
print(f"Actual Spend: ₹{actual_spend:,.2f}")
print(f"Best-Fit Spend: ₹{best_fit_total:,.2f}")
print(f"TRUE PROFIT RECOVERED: ₹{real_savings:,.2f}")

```
